```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Rango de Qp
Qp = np.linspace(0.1, 5, 2000)

# Coeficientes de la ecuación en x = Omega_H^2
a = 1
b = (1 / Qp**2) - 2
c = -1

# Solución positiva
x = (-b + np.sqrt(b**2 - 4*a*c)) / (2*a)
omega_H = np.sqrt(x)

# Valores especiales
Qp_1 = 1 / np.sqrt(2)
Qp_2 = 1 / np.sqrt(3)

def omegaH(Q):
    b = (1 / Q**2) - 2
    x = (-b + np.sqrt(b**2 + 4)) / 2
    return np.sqrt(x)

omegaH_1 = omegaH(Qp_1)
omegaH_2 = omegaH(Qp_2)

# ---- Gráfica ----
plt.figure()
plt.plot(Qp, omega_H, label=r'$\Omega_H(Q_p)$')

# Puntos
plt.plot(Qp_1, omegaH_1, 'o')
plt.plot(Qp_2, omegaH_2, 'o')

# Líneas de referencia
plt.axvline(Qp_1, linestyle='--')
plt.axvline(Qp_2, linestyle='--')

# Anotaciones espaciadas y claras
plt.annotate(
    rf'$Q_p=\frac{{1}}{{\sqrt{{2}}}}$' + '\n' +
    rf'$\Omega_H={omegaH_1:.3f}$',
    xy=(Qp_1, omegaH_1),
    xytext=(-120, 30),
    textcoords='offset points',
    arrowprops=dict(arrowstyle='->'),
    ha='right'
)

plt.annotate(
    rf'$Q_p=\frac{{1}}{{\sqrt{{3}}}}$' + '\n' +
    rf'$\Omega_H={omegaH_2:.3f}$',
    xy=(Qp_2, omegaH_2),
    xytext=(50, -50),
    textcoords='offset points',
    arrowprops=dict(arrowstyle='->'),
    ha='left'
)

plt.xlabel(r'$Q_p$')
plt.ylabel(r'$\Omega_H$')
plt.title(r'$\Omega_H$ en función de $Q_p$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Rango de Qp
Qp = np.linspace(0.1, 5, 2000)

# Coeficientes en x = Omega_G^2
a = 1
b = 1 / Qp**2
c = -1

# Solución positiva de la ecuación
x = (-b + np.sqrt(b**2 - 4*a*c)) / (2*a)
omega_G = np.sqrt(x)

# Valores especiales
Qp_1 = 1 / np.sqrt(2)
Qp_2 = 1 / np.sqrt(3)

def omegaG(Q):
    b = 1 / Q**2
    x = (-b + np.sqrt(b**2 + 4)) / 2
    return np.sqrt(x)

omegaG_1 = omegaG(Qp_1)
omegaG_2 = omegaG(Qp_2)

# ---- Gráfica ----
plt.figure()
plt.plot(Qp, omega_G, label=r'$\Omega_G(Q_p)$')

# Puntos
plt.plot(Qp_1, omegaG_1, 'o')
plt.plot(Qp_2, omegaG_2, 'o')

# Líneas verticales
plt.axvline(Qp_1, linestyle='--')
plt.axvline(Qp_2, linestyle='--')

# Anotaciones espaciadas y con flecha
plt.annotate(
    rf'$Q_p=\frac{{1}}{{\sqrt{{2}}}}$' + '\n' +
    rf'$\Omega_G={omegaG_1:.3f}$',
    xy=(Qp_1, omegaG_1),
    xytext=(-120, 30),
    textcoords='offset points',
    arrowprops=dict(arrowstyle='->'),
    ha='right'
)

plt.annotate(
    rf'$Q_p=\frac{{1}}{{\sqrt{{3}}}}$' + '\n' +
    rf'$\Omega_G={omegaG_2:.3f}$',
    xy=(Qp_2, omegaG_2),
    xytext=(50, -50),
    textcoords='offset points',
    arrowprops=dict(arrowstyle='->'),
    ha='left'
)

plt.xlabel(r'$Q_p$')
plt.ylabel(r'$\Omega_G$')
plt.title(r'$\Omega_G$ en función de $Q_p$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Rango de Qp
Qp = np.linspace(0.1, 5, 2000)

# ---- Omega_G(Qp) ----
b = 1 / Qp**2
x = (-b + np.sqrt(b**2 + 4)) / 2
omega_G = np.sqrt(x)

# ---- Margen de fase ----
Mphi = 90 - np.degrees(np.arctan(omega_G * Qp))

# Valores particulares
Qp_1 = 1 / np.sqrt(2)
Qp_2 = 1 / np.sqrt(3)

def omegaG(Q):
    b = 1 / Q**2
    x = (-b + np.sqrt(b**2 + 4)) / 2
    return np.sqrt(x)

Mphi_1 = 90 - np.degrees(np.arctan(omegaG(Qp_1) * Qp_1))
Mphi_2 = 90 - np.degrees(np.arctan(omegaG(Qp_2) * Qp_2))

# ---- Gráfica ----
plt.figure()
plt.plot(Qp, Mphi, label=r'$M_\phi(Q_p)$')

# Puntos
plt.plot(Qp_1, Mphi_1, 'o')
plt.plot(Qp_2, Mphi_2, 'o')

# Líneas de referencia
plt.axvline(Qp_1, linestyle='--')
plt.axvline(Qp_2, linestyle='--')
plt.axhline(Mphi_1, linestyle='--')
plt.axhline(Mphi_2, linestyle='--')

# Anotaciones (raw f-strings → SIN warnings)
plt.annotate(
    rf'$Q_p=\frac{{1}}{{\sqrt{{2}}}}$' + '\n' +
    rf'$M_\phi={Mphi_1:.1f}^\circ$',
    xy=(Qp_1, Mphi_1),
    xytext=(-120, 30),
    textcoords='offset points',
    arrowprops=dict(arrowstyle='->'),
    ha='right'
)

plt.annotate(
    rf'$Q_p=\frac{{1}}{{\sqrt{{3}}}}$' + '\n' +
    rf'$M_\phi={Mphi_2:.1f}^\circ$',
    xy=(Qp_2, Mphi_2),
    xytext=(40, -60),
    textcoords='offset points',
    arrowprops=dict(arrowstyle='->'),
    ha='left'
)

plt.xlabel(r'$Q_p$')
plt.ylabel(r'Margen de fase $M_\phi$ [°]')
plt.title(r'Margen de fase $M_\phi$ en función de $Q_p$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Frecuencia (rad/s)
w = np.logspace(-2, 2, 2000)

# Valores de Q a graficar
Q_values = [10, 1, 0.7, 0.2]

plt.figure(figsize=(10, 6))

# -------- Magnitud --------
plt.subplot(2, 1, 1)

for Q in Q_values:
    s = 1j * w
    H = 1 / (s**2 + (1/Q)*s + 1)
    mag = 20 * np.log10(np.abs(H))
    plt.semilogx(w, mag, label=rf'$Q={Q}$')

plt.ylabel('Magnitude (dB)')
plt.title('Bode Diagram')
plt.grid(True, which='both', linestyle='--')
plt.legend()

# -------- Fase --------
plt.subplot(2, 1, 2)

for Q in Q_values:
    s = 1j * w
    H = 1 / (s**2 + (1/Q)*s + 1)
    phase = np.angle(H, deg=True)
    plt.semilogx(w, phase)

plt.ylabel('Phase (deg)')
plt.xlabel('Frequency (rad/sec)')
plt.grid(True, which='both', linestyle='--')

plt.tight_layout()
plt.show()
```
